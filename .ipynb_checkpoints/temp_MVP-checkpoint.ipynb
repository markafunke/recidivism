```python
def cross_val_auc_all(model, X, y, rand = 122):
    
    # Prepare data for cross validation
    X, y = np.array(X), np.array(y)
    kf = KFold(n_splits=5, shuffle=True, random_state = rand)
    
    # Initiate lists to store results
    cv_AUCs = []
    
    for train_ind, val_ind in kf.split(X,y):
        X_train, y_train = X[train_ind], y[train_ind]
        X_val, y_val = X[val_ind], y[val_ind]
        
        # Create Classifier Object
#        model = classifier
        
        # linear model
        model.fit(X_train, y_train)
        y_pred = model.predict_proba(X_val)[:,1]
        cv_AUCs.append(metrics.roc_auc_score(y_val, y_pred))
        
        # Feature scaling
        # Only used by KNN
        #scaler = StandardScaler()
        #X_train_scaled = scaler.fit_transform(X_train)
        #X_val_scaled = scaler.transform(X_val)
    model_AUCs = round(np.mean(cv_AUCs),3)
    print(f"Model AUC: {model_AUCs}")
```

```python
# compas decile scores test -- > LOW
y_actual = compas_df[["two_year_recid"]]
compas_df["compas_guess"] = compas_df["score_text"].apply(lambda x: 0 if x == "Low" else 1)
y_pred = compas_df[["compas_guess"]]
metrics.confusion_matrix(y_actual,y_pred)
metrics.accuracy_score(y_actual, y_pred), metrics.precision_score(y_actual, y_pred), metrics.recall_score(y_actual, y_pred)
```

```python
# compas decile scores test --> HIGH
y_actual = compas_df[["two_year_recid"]]
compas_df["compas_guess"] = compas_df["score_text"].apply(lambda x: 1 if x == "High" else 0)
y_pred = compas_df[["compas_guess"]]
metrics.confusion_matrix(y_actual,y_pred)
metrics.accuracy_score(y_actual, y_pred), metrics.precision_score(y_actual, y_pred), metrics.recall_score(y_actual, y_pred)
```

```python
# compas decile scores test -- > 50%
y_actual = compas_df[["two_year_recid"]]
compas_df["compas_guess"] = compas_df["decile_score"].apply(lambda x: 1 if x > 5 else 0)
y_pred = compas_df[["compas_guess"]]
metrics.confusion_matrix(y_actual,y_pred)
metrics.accuracy_score(y_actual, y_pred), metrics.precision_score(y_actual, y_pred), metrics.recall_score(y_actual, y_pred)
```

```python
# mvp
X_train, X_val, y_train, y_val = train_test_split(X[["age","juv_misd_count"]] , y, test_size=0.2, random_state=4444)
```

```python
std = StandardScaler()

X_train_scaled = std.fit_transform(X_train)
X_val_scaled = std.transform(X_val)

lr1 = LogisticRegression()
lr1.fit(X_train_scaled, y_train)
```

```python
lr1_pred = lr1.predict(X_val_scaled)  # This is a hard classifier
(metrics.accuracy_score(y_val, lr1_pred), metrics.precision_score(y_val, lr1_pred)
,metrics.recall_score(y_val, lr1_pred))
```

```python
THRESHOLD = .5
lr1_pred = np.where(lr1.predict_proba(X_val_scaled)[:,1] > THRESHOLD, 1, 0)
(metrics.accuracy_score(y_val, lr1_pred), metrics.precision_score(y_val, lr1_pred)
,metrics.recall_score(y_val, lr1_pred))
```

```python
df_EDA = X
df_EDA["y"] = y

mask = df_EDA.juv_misd_count > -1
df_EDA_viz = df_EDA[mask]
df_EDA_viz["juv_m_gt1"] = df_EDA_viz["juv_misd_count"].apply(lambda x: 0 if x == 0 else 1)
df_EDA_viz["juv_f_gt1"] = df_EDA_viz["juv_fel_count"].apply(lambda x: 0 if x == 0 else 1)
```

```python
# barplot of marriage rating grouped by affair (True or False)
pd.crosstab(df_EDA_viz.juv_m_gt1, df_EDA_viz.y.astype(bool)).plot(kind='bar')
plt.title('Marriage Rating Distribution by Affair Status')
plt.xlabel('Marriage Rating')
plt.ylabel('Frequency')
```

```python
affair_yrs_married = pd.crosstab(df_EDA_viz.juv_m_gt1, df_EDA_viz.y)
affair_yrs_married.div(affair_yrs_married.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Reoffending % by Juvenile Crime Count')
plt.xlabel('Juvenile Priors')
plt.ylabel('Percentage')
```
